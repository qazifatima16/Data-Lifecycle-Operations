# Fetching data from API's

This project fetches the data from the API of TMDB website. The data contains various attributes about high rated data. We fetch the following attributes - popularity, vote_count, original_language, title, vote_average, overview, release_date. 500 pages are to be fetched from the API and each page had 20 movies

Note: I initially started with this project and upon fetching and observing the data I realized that this is a clean dataset and dosen't need much cleaning ,so I switched to another project "Country.py". "Country.py" project reads data directly from a csv file.

```python
#importing libraries
import requests
import pandas as pd
import socket
import urllib.request
import urllib.error
```

```python
#creating empty dataframe
#this pice of code may take some time to execute
final=pd.DataFrame()

#fetching each page from the 500 pages in API
for i in range(1,501):
    try:
        #getting data from a page
        response=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i)).json()
        popularity=[]
        vote_count=[]
        original_language=[]
        title=[]
        vote_average=[]
        overview=[]
        release_date=[]

        #appending the attribute value to the respective  list
        for movie in response['results']:

            popularity.append(movie['popularity'])
            vote_count.append(movie['vote_count'])
            original_language.append(movie['original_language'])
            title.append(movie['title'])
            vote_average.append(movie['vote_average'])
            overview.append(movie['overview'])
            release_date.append(movie['release_date'])

        #creating a dictionary with attribute name and list of attribute values as key/value pairs
        d={'title':title, 'overview':overview, 'original_language':original_language, 'release_date':release_date, 'popularity':popularity, 'vote_count':vote_count, 'vote_average':vote_average}
        df=pd.DataFrame(d)
    
        #appending the data genetated by each page in 'final' dataframe
        final=final.append(df,ignore_index=True)
        
    #handling errors
    except ConnectionResetError:
        continue
    except urllib.request.HTTPError:
        continue
    except TimeoutError:
        continue
    except urllib.error.URLError:
        continue
    except socket.error:
        continue
        
```

```python
#saving the data as a csv file
final.to_csv('movies.csv')
```
